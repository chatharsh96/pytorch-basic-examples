```python
import pandas as pd
```

```python
df = pd.read_csv('YearPredictionMSD.txt', header='None')
    
```

```python
cols = df.columns
numcols = df._get_numeric_data().columns
a = list(set(cols) - set(numcols))
a
```

```python
df.isnull().sum()
```

```python
data = df
```

```python
outliers = {}
for column in data.columns:
    min_t = data[column].mean() - (3 * data[column].std())
    max_t = data[column].mean() + (3 * data[column].std())
    
    count = ((data[column] < min_t) | (data[column] > max_t)).sum()
    percentage = count / data.shape[0]
    outliers[column] = f"{percentage:.3f}"

print(outliers)
```

```python
X = data.iloc[:,1:]
Y = data.iloc[:,0]
```

```python
X = (X - X.mean())/X.std()
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_shuffle = X.sample(frac=1,random_state=0)
y_shuffle = Y.sample(frac=1,random_state=0)
```

```python
x_new, x_test, y_new, y_test = train_test_split(x_shuffle, y_shuffle, test_size=0.2, random_state=0)

val_per = x_test.shape[0]/x_new.shape[0]

x_train, x_val, y_train, y_val = train_test_split(x_new, y_new, test_size=val_per, random_state=0)
```

```python
print(x_train.shape, y_train.shape)
print(x_val.shape, y_val.shape)
print(x_test.shape, y_test.shape)
```

```python
import torch
import torch.nn as nn
print("Number of GPU: ", torch.cuda.device_count())
print("GPU Name: ", torch.cuda.get_device_name())


device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('Using device:', device)
```

```python
x_train = torch.tensor(x_train.values).float().cuda()
y_train = torch.tensor(y_train.values).float().cuda()

x_val = torch.tensor(x_val.values).float().cuda()
y_val = torch.tensor(y_val.values).float().cuda()

x_test = torch.tensor(x_test.values).float().cuda()
y_test = torch.tensor(y_test.values).float().cuda()
```

```python
print(type(x_val))
print(type(y_val))
print(type(x_test))
print(type(y_test))
```

```python
model = nn.Sequential(nn.Linear(x_train.shape[1], 200), nn.ReLU(), nn.Linear(200,100), nn.ReLU(), nn.Linear(100, 50), nn.ReLU(), nn.Linear(50,1)).cuda()
```

```python
model
```

```python
loss_function = torch.nn.MSELoss()
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
for i in range(10000):
    y_pred = model(x_train).squeeze()
    loss = loss_function(y_pred, y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if i%100 == 0:
        print(i, loss.item())
```

```python
pred = model(x_test[0])
print("ground truth:", y_test[0].item(), "Prediction:", pred.item())
```
