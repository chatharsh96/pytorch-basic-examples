```python
import pandas as pd
```

```python
data = pd.read_excel('default of credit card clients.xls', skiprows=1)
data.head()
data
```

```python
# remove not relevant features

dataclean = data.drop(columns=["ID", "SEX"])
dataclean
```

```python
# check for empty/missing values

total = dataclean.isnull().sum()
percent = (dataclean.isnull().sum()/dataclean.isnull().count()*100)
pd.concat([total, percent], axis=1, keys=['Total', 'Percent']).transpose()
```

```python
# check for outliers

outliers = {}
for column in data.columns:
    min_t = data[column].mean() - (3 * data[column].std())
    max_t = data[column].mean() + (3 * data[column].std())
    
    count = ((data[column] < min_t) | (data[column] > max_t)).sum()
    percentage = count / data.shape[0]
    outliers[column] = f"{percentage:.3f}"

print(outliers)
```

```python
# detecting class imbalance

target = dataclean["default payment next month"]
yes = target[target == 1].count()
no = target[target == 0].count()

print("yescount & yespercentage:", yes, str(yes/len(target)*100))
print("nocount & nopercentage:", no, str(no/len(target)*100))
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,5))
plt.bar("yes", yes)
plt.bar("no", no)
ax.set_yticks([yes, no])
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
datayes = dataclean[dataclean["default payment next month"] == 1]
data_no = dataclean[dataclean["default payment next month"] == 0]

oversampling  = datayes.sample(no, replace=True, random_state=0)
data_resampled = pd.concat([data_no, oversampling], axis=0)
data_resampled
```

```python
target2 = data_resampled["default payment next month"]
yes = target2[target2 == 1].count()
no = target2[target2 == 0].count()

print("yescount & yespercentage:", yes, str(yes/len(target2)*100))
print("nocount & nopercentage:", no, str(no/len(target2)*100))
```

```python
# Split the dataset into features matrix and target matrix

data_resampled = data_resampled.reset_index(drop=True)
x = data_resampled.drop(columns=["default payment next month"])
y = data_resampled["default payment next month"]
# print(x)
print(y)
```

```python
# rescale the values of features matrix, to avoid bias

x = (x - x.min())/(x.max() - x.min())
x.head()
```

```python
finaldata = pd.concat([x,y], axis=1)
finaldata.to_csv("finalprepareddata.csv", index=False)
```
